4. Par ou Ímpar
Peça um número inteiro. Mostre "Número par" ou
"Número ímpar". (Use if e else).

```python
#Exercício 04;

numero4 = float(input("Insira á seguir um número: "))

if numero4 % 2 == 0:
    print("O número inserido é Par.")

else:
    print("O número inserido é Ímpar.")
```

7. Classificação de idade
Peça a idade da pessoa. Mostre:
▶ Criança (0–12)
▶ Adolescente (13–17)
▶ Adulto (18 ou mais)
(Use if, elif e else).

```python
#Exercício 07:

idade2 = int(input("Insira a idade do usuário:"))

if idade2 >= 0 and idade2 <=12:
    print("O usuário é Criança.")

elif idade2 >= 13 and idade2 <= 17:
    print("O usuário é Adolescente.")

else:
    print("O usuário é Adulto")
    
```

8. Nota do aluno
Peça a nota do aluno. Mostre:
▶ Aprovado (nota ≥ 7)
▶ Recuperação (nota entre 5 e 6)
▶ Reprovado (nota < 5)
(Use if, elif e else).

```python
#Exercício 08:

nota2 = float(input("Insira a nota do aluno:"))

if nota2 >= 7:
    print("O aluno está Aprovado.")

elif nota2 >= 5 and nota2 <= 6:
    print("O aluno está de Recuperação")

else:
    print("O aluno está Reprovado")
    
```

10. Sistema simples de login
Peça ao usuário nome e senha. Considere:
▶ usuario_correto = "admin"
▶ senha_correta = "1234"
Se os dados estiverem corretos, mostre "Login
realizado com sucesso". Caso contrário, "Usuário ou
senha incorretos". (Use o operador ==).

```python
#Exercício 10

usuario = str(input("Insira o nome do usuário ou login: "))
senha = int(input("Insira a senha do login: "))

if usuario == "admin" and senha == 1234:
    print("Login realizazdo com sucesso!!")

else:
    print("Usuário ou senhas incorretos.")
```

Desafio Extra

Crie um programa que peça um número e mostre:
▶ Número positivo
▶ Número negativo
▶ Número igual a zero
Dica: Use as estruturas if, elif e else para resolver
esse problema.

Prof. Me. Ujeverson Tavares

```python
#DESAFIO EXTRA

numero_extra = float(input("Insira á seguir um número: "))

if numero_extra > 0:
    print("O número inserido é POSITIVO.")

elif numero_extra < 0: 
    print("O número inserido é NEGATIVO.")

else:
    print("O número inserido é 0.")
```
