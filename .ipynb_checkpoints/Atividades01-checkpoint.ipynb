Exercício 3

Crie duas variáveis: nome_do_aluno e idade_do_aluno.
Depois mostre na tela algo assim:

Meu nome é Ana e tenho 18 anos

Dica: Use vírgula dentro do print().

```python
#EXERCÍCIO 03;

nome_do_aluno = "Ana"
idade_do_aluno = 18

print(f"Meu nome é {nome_do_aluno} e tenho {idade_do_aluno} anos de idade")
```

```python
Exercício 4 (Desafio Extra)

Crie três variáveis: cidade, estado e pais. E mostre:
Eu moro em Salvador - BA, Brasil

Dica: Use concatenação com +.
```

```python
#EXERCÍCIO 04;

cidade = "Goiânia"
estado = "Goiás (GO)"
pais = "Brasil"

endereço_completo = "Eu moro em " + cidade + " - " + estado + ", " + pais
print(endereço_completo)
```
